```python
#read in data
base = '/Users/adarekar/Documents/College/Junior/DSC 495/T_ONTIME_REPORTING-'

jan = pd.read_csv(base + '01-2020.csv')
feb = pd.read_csv(base + '02-2020.csv')
mar = pd.read_csv(base + '03-2020.csv')
apr = pd.read_csv(base + '04-2020.csv')
may = pd.read_csv(base + '05-2020.csv')
jun = pd.read_csv(base + '06-2020.csv')
jul = pd.read_csv(base + '07-2020.csv')
aug = pd.read_csv(base + '08-2020.csv')
sept = pd.read_csv(base + '09-2020.csv')
octo = pd.read_csv(base + '10-2020.csv')
nov = pd.read_csv(base + '11-2020.csv')
dec = pd.read_csv(base + '12-2020.csv')
```

## Data Cleaning

```python
#combine all datasets
flight_delays = pd.concat([jan, feb, mar, apr, may, jun, jul, aug, sept, octo, nov, dec])
flight_delays
```

```python
#create variable for total delay in minutes
#create variable for name of month 
tot_delay_arr = []
tot_delay = 0 
month_name = []
delay_type  =[]
for index, row in flight_delays.iterrows(): 
    tot_delay = row['CARRIER_DELAY'] + row['WEATHER_DELAY'] + row['NAS_DELAY'] + row['SECURITY_DELAY'] + row['LATE_AIRCRAFT_DELAY']
    tot_delay_arr.append(tot_delay)
    
    if row['MONTH'] == 1: 
        month_name.append('January')
    elif row['MONTH'] == 2: 
        month_name.append('February')
    elif row['MONTH'] == 3: 
        month_name.append('March')
    elif row['MONTH'] == 4: 
        month_name.append('April')
    elif row['MONTH'] == 5: 
        month_name.append('May')
    elif row['MONTH'] == 6: 
        month_name.append('June')
    elif row['MONTH'] == 7: 
        month_name.append('July')
    elif row['MONTH'] == 8: 
        month_name.append('August')
    elif row['MONTH'] == 9: 
        month_name.append('September')
    elif row['MONTH'] == 10: 
        month_name.append('October')
    elif row['MONTH'] == 11: 
        month_name.append('November')
    elif row['MONTH'] == 12: 
        month_name.append('December')
        
    
flight_delays['TOT_DELAY'] = tot_delay_arr
flight_delays['MONTH_NAME'] = month_name
flight_delays
```
